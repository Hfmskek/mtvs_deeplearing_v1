```python
# 1. 필요 모듈 불러오기
from bs4 import BeautifulSoup
import requests
```

```python
# 2. 페이지 요소 가져오기 웹페이지보다 모바일 화면으로 가져오면 더 잘 됨.
blog_url="https://m.blog.naver.com/allthetime_/223317346734"
page = requests.get(blog_url)
```

```python
# 3. Beautifulsoup 을 이용해서 파싱
bs_result = BeautifulSoup(page.text, "html.parser")
bs_result
```

```python
# 4. 요소 추출
result = bs_result.find_all("span", class_="se-fs-")
```

```python
result_text=""
for item in result:
    result_text += item.text
```
