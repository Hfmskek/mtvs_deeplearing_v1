```python
# 1. 필요 모듈 불러오기
from bs4 import BeautifulSoup
import requests
```

```python
# 2. html 요소 열기
keyword = "구미"
url = f"https://search.naver.com/search.naver?sm=tab_hty.top&where=nexearch&ssc=tab.nx.all&query={keyword}+날씨"
page=requests.get(url)
```

```python
# 3. Beautifulsoup 을 이용해서 파싱
bs_result = BeautifulSoup(page.text, "html.parser")
bs_result
```

## Copy selector 를 이용하여 가져오기

```python
# 4. 기본 요소 추출
result_list = [] # 결과물을 저장할 리스트 생성
copy_selector = "#main_pack > section.sc_new.cs_weather_new._cs_weather > div > div:nth-child(1) > div.content_wrap > div.open > div:nth-child(1) > div > div.weather_info > div > div._today > div.weather_graphic > div.temperature_text > strong"
copy_data = bs_result.select_one(copy_selector)
copy_text = copy_data.get_text(separator=",")
result_list.append(copy_text)
result_list
```

### 두 번에 걸쳐서 가져오기 - 번외

```python
today_temperature = bs_result.find("div", class_="_today")
data = today_temperature.find("div", class_="temperature_text")
data
```

```python
now_temperature = data.get_text(separator=",")
now_temperature
```

```python
temperature = bs_result.find_all("div", class_="temperature_text")
temperature
```

```python
now_temperature = temperature[0].get_text(separator=",")
now_temperature
```

```python
# 5. 날씨 요소 추출
weather = bs_result.find("span", class_="weather")
weather_result = weather.text
result_list.append(weather_result)
result_list
```
