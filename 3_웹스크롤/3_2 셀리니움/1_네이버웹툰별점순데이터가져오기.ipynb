```python
# 1. 모듈 임포트
import time
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
```

```python
# 2. 드라이버 옵션 설정
chrome_options = webdriver.ChromeOptions()
chrome_options.add_experimental_option('detach', True)
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()),
                          options=chrome_options)
```

```python
# 3. 웹브라우저에 주소 넣기
url = 'https://comic.naver.com/webtoon'
driver.get(url)
time.sleep(2)
```

```python
buttons_list= driver.find_elements(By.CLASS_NAME, )
```

```python
# 4. 별점 순으로 이동
css_select = "#container > div.component_wrap.type2 > div.ComponentHead__component_head--O2tPr > div > div > button:nth-child(4)"
driver.find_element(By.CSS_SELECTOR, css_select).click()
time.sleep(1)
```

```python
def solution(numbers, our_score, score_list):
    answer = []
    for i in range(len(numbers)):
        if numbers[score_list[i]] == our_score[i]:
            answer.append("Same")
        else:
            answer.append("Different")
    
    return answer
```
