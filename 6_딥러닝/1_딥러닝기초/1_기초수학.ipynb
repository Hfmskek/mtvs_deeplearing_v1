#### 1차함수 그려보기
- $y = ax + b$
    - $a: 기울기$
    - $b: y 절편$


```python
def linear_fc(x):
    a = 2
    b = 3
    fx = a*x + b
    return fx
```

#### 2차함수 그려보기
- $y = ax^2 + bx + c$

```python
def quadratic_fc(x):
    a = 2
    b = 3
    c = 4
    fx = (a*x**2) + (b*x) + c
    return fx
```

```python
# 2차함수 그려보기
def display_f(title, fx):
    x = np.arange(-3, 3, 0.1)
    y = fx(x)
    plt.figure(figsize=(6,5))
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel("y", rotation=0)
    plt.title(title)
    plt.show()
```

```python
display_f("이차함수", quadratic_fc)
```

#### 다항함수 그려보기
- $y = ax^3 + bx^2 + cx + d$
    - $a, b, c : 계수$
    - $d: y 절편$

```python
def polynomial_f(x):
    a = 2
    b = 1
    c = -7
    d = 1
    fx = (a*x**3) + (b*x**2) + (c*x) + d
    return fx
```

```python
display_f("다항함수", polynomial_f)
```

### 구간내의 최소값 구해보기

```python
def get_minimum(x1, x2, f):
    x = np.arange(x1, x2, 0.01)
    y = f(x)
    minimum_y = np.round(np.min(y), 2)
    minimum_x = np.round(x[np.argmin(y)], 2)
    return minimum_x, minimum_y
```

```python
get_minimum(-1, 2, polynomial_f)
```

### 지수함수 그려보기

### 로그함수 그려보기
- $y = \log_a(x) $ : 밑이 a 인 로그함수
- $y = \log_10(x) $ : 밑이 10인 로그함수(상용로그) 10 생략 가능 $log(x)$
- $y = \ln(x) $ : 밑이 e 인 로그함수(자연로그)


```python
# 밑이 2인 로그함수 그려보기
def log_2f(x):
    fx = np.log2(x)
    return fx
```

```python
# 밑이 10인 로그함수 그려보기
def log_10f(x):
    fx = np.log10(x)
    return fx
```

```python
# 밑이 e인 로그함수 그려보기
def log_ef(x):
    fx = np.log(x)
    return fx
```

```python
log_fx_list = [log_2f, log_10f, log_ef]
```

```python
x = np.arange(0.05, 5, 0.01)

plt.figure(figsize=(10, 6))

for fx in log_fx_list:
    y = fx(x)
    plt.plot(x, y, label=fx.__name__)

plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.title("로그함수 전체 보기")
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.text(1, -0.5, (1,0), fontsize=12)
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```

### Logistic Regression 을 위한 함수 
시그모이드 함수
- $y=\frac{1}{(1+e^{-z})}$
- 주로 이진 분류 문제에서 출력층의 활성화 함수로 사용됨
- 앞으로 사용할 신경망의 마지막 층에서 이진분류로 사용될 예정
- $x=0$일 경우 0.5을 출력
- $x=10$일 경우 1을 출력 
- $x=-10$일 경우 0을 출력. 즉, 양수일 때는 1, 음수일 때는 0을 출력
- 단점 : 미분시에 ""기울기 소실"" 이라는 문제가 생길 수 있음(기울기 소실에 대해 검색)

```python
#  시그모이드 함수 정의
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# z 값을 범위 설정
z = np.linspace(-10, 10, 100)
# 시그모이드 함수 적용
y = sigmoid(z)

# 차트 그리기
plt.figure(figsize=(10, 6))
plt.plot(z, y, label="시그모이드 함수")
plt.xlabel('z')
plt.ylabel('sigmoid(z)')
plt.title('Sigmoid 시그모이드')

# 축에 선 그리기
plt.axhline(y=0, color='k', linestyle='--')
plt.axvline(x=0, color='k', linestyle='--')
plt.legend()
plt.grid(True)
plt.show()
```

### 미분에 대해 알아보기
<div style="line-height: 2;">

- 어떤 순간의 변화율을 나타내는 것
- 기울기란?
    - 어떤 함수를 나타내는 그래프에서 특정한 점의 미분값을 구하는 것은 해당 점에서의 점선의 기울기를 뜻함. 
    - 최소 지점일 때에는 미분값이 0이 되는 지점
- 미분에 대한 실생활 예)
    1. 자동차의 순간 속도
        - 시간에 대한 거리의 변화량이 속도
        - 평균 속도라는 것은 $\frac{총이동거리}{총시간}$
        - 순간 속도라는 것은 그 순간의 속도 : 차량의 계기판에 표시되는 속도
            - 이동거리를 시간에 대해 미분 : $v(t) = \frac{d(5t^2)}{dt} = 10t$
    2. 비용 함수에서의 손실함수 계산
        - 전체 비용 함수(MSE) :   $MSE = (W - 2)^2 + (B - 3)^2$
        - W에 대한 편미분:
            - $\frac{\partial MSE}{\partial W} = 2(W - 2)$
            - $2(W - 2) = 0$
            - $W = 0$
        - B에 대한 편미분:
            - $\frac{\partial MSE}{\partial B} = 2(B - 3)$
            - $2(B - 2) = 0$
            - $B = 0$
    3. 다양한 w 값에 따른 손실함수 계산
        - 비용 함수 $MSE$는 다음과 같이 정의될때 :
            - $MSE = (w_1 - 2)^2 + (w_2 - 3)^2 + (w_3 - 4)^2$
            - $\frac{\partial MSE}{\partial w_1} = 2(w_1 - 2) = 0, \; w_1 = 2$
            - $\frac{\partial MSE}{\partial w_2} = 2(w_2 - 3) = 0, \; w_2 = 3$
            - $\frac{\partial MSE}{\partial w_3} = 2(w_3 - 4) = 0, \; w_3 = 4$
</div>

```python
from mpl_toolkits.mplot3d import Axes3D

# 가중치(w)와 절편(b)의 범위 설정
w_range = np.linspace(0, 4, 100)
b_range = np.linspace(0, 6, 100)
W, B = np.meshgrid(w_range, b_range)

# 비용 함수 계산
MSE = (W - 2)**2 + (B - 3)**2

# 비용 함수 그래프
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(W, B, MSE, cmap='coolwarm')

# 최소 지점 표시
min_w, min_b = 2, 3
ax.scatter(min_w, min_b, 0, color='red', s=100, label='최소 지점 : (w, b) = (2, 3)')
ax.legend()

# 축 레이블 설정
ax.set_xlabel('Weight (w)')
ax.set_ylabel('Bias (b)')
ax.set_zlabel('MSE')

# 뷰 시점 설정
ax.view_init(elev=10, azim=45)

plt.title('MSE 함수의 최소 비용에 대한 최소 지점')

plt.show()
```

### 미분 체험해보기
- 본래 함수
    - $f(x) = x^3 - 6x^2 + 11x - 6$
- 미분 함수
    - $f'(x) = 3x^2 - 12x + 11$

```python
from scipy.optimize import fsolve

# 계수 정의
a = 1
b = -4/3
c = -4
d = 0
e = 5

# 함수 정의
def f(x):
    return (a * x**4) + (b * x**3) + (c * x**2) + (d * x) + e

# 함수의 미분
def f_prime(x):
    return (4 * a * x**3) + (3 * b * x**2) + (2 * c * x) + d

# 극소값 찾기 (1차 미분이 0이 되는 지점)
initial_guesses = [-2, 5]
x0_1, x0_2 = fsolve(f_prime, initial_guesses)
y0_1 = f(x0_1)
y0_2 = f(x0_2)
slope_1 = f_prime(x0_1)
slope_2 = f_prime(x0_2)

# 접선의 방정식
def tangent_line(x, x0, y0, slope):
    return slope * (x - x0) + y0

# x 값의 범위 설정
x = np.linspace(-3, 5, 400)
y = f(x)
tangent_y_1 = tangent_line(x, x0_1, y0_1, slope_1)
tangent_y_2 = tangent_line(x, x0_2, y0_2, slope_2)

# 차트 그리기
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='$f(x) = x^4 - 6x^3 + 11x^2 - 6x$', color='black')
plt.plot(x, tangent_y_1, label=f'접선 $y = {y0_1:.2f}$ (x = {x0_1:.2f})', linestyle='--')
plt.plot(x, tangent_y_2, label=f'접선 $y = {y0_2:.2f}$ (x = {x0_2:.2f})', linestyle='--')
plt.scatter([x0_1, x0_2], [y0_1, y0_2], color='red', zorder=5)
plt.text(x0_1, y0_1-1.5, f'({x0_1:.2f}, {y0_1:.2f})', fontsize=12, verticalalignment='bottom')
plt.text(x0_2, y0_2-1.5, f'({x0_2:.2f}, {y0_2:.2f})', fontsize=12, verticalalignment='bottom')

# x = 3에서의 접선 구하기
x0_3 = 2.3
y0_3 = f(x0_3)
slope_3 = f_prime(x0_3)
tangent_y_3 = tangent_line(x, x0_3, y0_3, slope_3)
plt.plot(x, tangent_y_3, label=f'최소가 아닐때 $y = {slope_3:.2f}(x - {x0_3:.2f}) + {y0_3:.2f}$ (x = {x0_3:.2f})', linestyle=(0, (3, 1, 1, 1)), color='b', linewidth=2)
plt.text(x0_3+0.2, y0_3, f'({x0_3:.2f}, {y0_3:.2f})', fontsize=12, verticalalignment='bottom')
plt.scatter([x0_3],[y0_3], color='blue', zorder=5)

plt.xlabel('$x$')
plt.ylabel('$y$', rotation=0)
plt.ylim(-15, 10)  # y축 범위 설정
plt.title('4차 함수와 두 개의 최소점에서의 접선')
plt.legend(loc='lower left')
plt.grid(True)
plt.show()
```
