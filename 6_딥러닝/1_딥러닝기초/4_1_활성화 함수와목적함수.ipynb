### 계단함수

```python
import numpy as np
import matplotlib.pyplot as plt

def step_function(x):
    return np.where(x >= 0, 1, 0)

# 입력 값 범위 설정
x = np.linspace(-10, 10, 100)
y_step = step_function(x)

# 시각화
plt.plot(x, y_step, label='Step Function')
plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

### 시그모이드 함수

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

y_sigmoid = sigmoid(x)

# 시각화
plt.plot(x, y_sigmoid, label='Sigmoid Function')
plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

### 렐루 함수

```python
def relu(x):
    return np.maximum(0, x)

y_relu = relu(x)

# 시각화
plt.plot(x, y_relu, label='ReLU Function')
plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

### 탄젠트 함수

```python
def tanh(x):
    return np.tanh(x)

y_tanh = tanh(x)

# 시각화
plt.plot(x, y_tanh, label='Tanh Function')
plt.title("Activation Functions")
plt.xlabel("Input")
plt.ylabel("Output")
plt.grid(True)
plt.legend()
plt.show()
```

### 소프트맥스 함수

```python
import koreanize_matplotlib

def softmax(x):
    e_x = np.exp(x - np.max(x))
    return e_x / e_x.sum(axis=0)

# 소프트맥스는 다차원 입력에서 사용
x_softmax = np.array([1.0, 2.0, 3.0])
y_softmax = softmax(x_softmax)

# 시각화
plt.bar(['class1', 'class2', 'class3'], y_softmax, label='소프트맥스 함수 결과')
plt.title("소프트맥스 함수")
plt.xlabel("클래스")
plt.ylabel("확률", rotation=0)
plt.grid(True)
plt.legend()
plt.show()
```

### 손실함수

```python
import numpy as np

# MSE 손실 함수
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# 교차 엔트로피 손실 함수
def cross_entropy_loss(y_true, y_pred):
    epsilon = 1e-15
    y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
    return -np.sum(y_true * np.log(y_pred))

# 예제 입력
y_true = np.array([1, 0, 1])
y_pred = np.array([0.9, 0.1, 0.8])

# 손실 계산
mse = mse_loss(y_true, y_pred)
cross_entropy = cross_entropy_loss(y_true, y_pred)

print("MSE 손실:", mse)
print("교차 엔트로피 손실:", cross_entropy)
```
