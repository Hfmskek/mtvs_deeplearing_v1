```python
from langchain_community.document_loaders import PyPDFLoader
from langchain_community.vectorstores import FAISS
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_chroma import Chroma
from langchain.retrievers.multi_vector import MultiVectorRetriever
```

```python
pdf_file_path = "소나기 - 황순원.pdf"
loader = PyPDFLoader(pdf_file_path)
rain_pages = loader.load()
pages=rain_pages
```

```python
pages
```

```python
for num, context in enumerate(pages):
    print(num, context.metadata)
```

```python
# 여러 개의 문서 다뤄보기
from glob import glob
file_list = glob("./*나무위키.pdf")
file_list
```

```python
docs = []

for file_path in file_list:
    loader = PyPDFLoader(file_path)
    pages = loader.load()
    docs += pages
docs
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size = 1000,
    chunk_overlap = 100
)
docs_spliter = text_splitter.split_documents(rain_pages)
docs_spliter[:10]
```

```python
from langchain_community.embeddings import OllamaEmbeddings

ollama_emb = OllamaEmbeddings(
    model="mxbai-embed-large",
    # model="llama3.1",
    num_gpu=1,
    show_progress=True
)
```

```python
# 벡터 저장소 경로
DB_PATH = "./chroma_db"
vectorstore = Chroma.from_documents(
    docs_spliter,
    ollama_emb,
    persist_directory=DB_PATH
)
```

```python
# 저장해놓은 벡터 저장소를 불러오기
my_db = Chroma(persist_directory=DB_PATH,
               embedding_function=ollama_emb)
```

```python
query = "소녀가 소년을 어떻게 생각하는지 알려줘?"
retriever = my_db.as_retriever(
    search_kwargs={"k":5}
)
docs = retriever.invoke(query)
docs
```

```python
def format_docs(docs):
    return "\n\n".join([doc.page_content for doc in docs])
```

```python
format_docs(docs)
```

```python
# !pip install langchainhub
```

```python
# 프롬프트 설정
from langchain import hub
rag_prompt = hub.pull("rlm/rag-prompt")
rag_prompt
```

```python
from langchain.chat_models import ChatOllama
model = ChatOllama(
    model="llama3.1",
    temperature=0.3
    )
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser

# 최종 RAG chain 만들어보기
rag_chain = (
    {"context": retriever | format_docs, "question": RunnablePassthrough()}
    | rag_prompt
    | model
    | StrOutputParser()
)
```

```python
query = "소녀가 소년을 어떻게 생각하는지 알려줘?"
response = rag_chain.stream(query)
for chunk in response:
    print(chunk, end="", flush=True)
```

```python
# You are an assistant for question-answering tasks. 
# Use the following pieces of retrieved context to answer the question. 
# If you don't know the answer, just say that you don't know. 
# Use three sentences maximum and keep the answer concise.
```

```python
from langchain.prompts import ChatPromptTemplate
# 새로 커스터마이즈한 프롬프트로 작성해보기
template = """
너는 질문에 대답을 잘 하는 어시스턴트야
아래의 검색된 Context 기반으로 Question 에 대답을 하도록 해
모르면 모른다고 대답을 하고
질문에 좀더 성의 있게 유추해서 대답을 할 수 있다면 하도록 해
한국어 반드시 답변 하도록 해
Question: {question} 
Context: {context} 
Answer:
"""
custom_prompt = ChatPromptTemplate.from_template(template=template)
```

```python
# 최종 RAG chain 만들어보기
rag_chain = (
    {"context": openai_retriever | format_docs, "question": RunnablePassthrough()}
    | custom_prompt
    | model
    | StrOutputParser()
)
query = "소녀가 소년을 어떻게 생각하는지 알려줘"
response = rag_chain.stream(query)
for chunk in response:
    print(chunk, end="", flush=True)
```
