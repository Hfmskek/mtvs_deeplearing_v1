```python
import re
from tqdm import tqdm

noun_result = []  #[ [] , [], [] ]

for temp in tqdm(df['text']):
    han = re.compile('[^ㄱ-ㅣ가-힣]')
    temp_result = han.sub(' ',temp) #문장에서 한글만 남기고 제거
    temp_noun = mecab.nouns(temp_result)
    result_noun = [ x for x in temp_noun if len(x) >1]
    noun_result.append(result_noun)

print(noun_result)
```

```python
# bag of word 벡터 생성
from sklearn.feature_extraction.text import CountVectorizer
count_vectorizer = CountVectorizer(tokenizer=lambda x:x, lowercase=False)
bow_vect = count_vectorizer.fit_transform(noun_result)
word_list = count_vectorizer.get_feature_names_out()
print(word_list)
len(word_list)
```

```python
#tf-idf생성
from sklearn.feature_extraction.text import TfidfTransformer

tf_vectorizer = TfidfTransformer()
tf_idf = tf_vectorizer.fit_transform(bow_vect)
tf_idf.toarray()
```

```python
# 감성분석
# 데이터 전처리
df['score_real']=np.where(df['score'] >= 4 , 1, 0)
df.head()
```

```python
df['score_real'].value_counts()
```

```python
from sklearn.model_selection import train_test_split

x = tf_idf
y = df['score_real']

X_train, X_test, y_train, y_test = train_test_split(x,y,test_size=0.3 , random_state=1,stratify=df['score_real'])
```

```python
print(X_train.shape)
print(X_test.shape)
```

```python
#모델학습 이진분류
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
lr_model = LogisticRegression(random_state=1)
lr_model.fit(X_train,y_train)
```

```python
y_pred = lr_model.predict(X_test)
y_pred
```

```python
accuracy_score(y_test,y_pred)
```

```python
precision_score(y_test,y_pred)
```

```python
def comment_predict(text):
    sentence_test = text
    han = re.compile('[^ㄱ-ㅣ가-힣]')
    temp_result = han.sub(' ',sentence_test)
    temp_noun = mecab.nouns(temp_result)
    result_noun = [x for x in temp_noun if len(x) > 1]
    vec_test = count_vectorizer.transform([result_noun])
    vec_test.toarray().sum()
    tf_test = tf_vectorizer.transform(vec_test)
    tf_test.toarray().sum()

    result = ''
    if lr_model.predict(tf_test)[0] == 1:
        result = '긍정'
    else :
        result = '부정'

    return result
```
