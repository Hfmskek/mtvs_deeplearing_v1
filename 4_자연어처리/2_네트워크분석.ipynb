```python
import pandas as pd
import koreanize_matplotlib
```

```python
df = pd.read_csv('appreply.csv')
reply_df = df.iloc[3:]['text']
reply_df
```

```python
from konlpy.tag import Mecab
from tqdm import tqdm

mecab = Mecab(dicpath=r"C:\mecab\mecab-ko-dic")

reply_list = []
for temp in tqdm(reply_df):
    words = mecab.nouns(temp)
    w_list = []
    for word in words:
        if len(word) > 1:
            w_list.append(word)

    reply_list.append(w_list)

reply_list
```

```python
!pip install apyori
```

```python
from apyori import apriori
result = list(apriori(reply_list,min_support=0.05))
result
```

```python
df = pd.DataFrame(result)
df
```

```python
item_len = []

for temp in df['items']:
    item_len.append(len(temp))

df['length'] = item_len
```

```python
df = df[(df['length']==2) & (df['support']>0.05)]
df
```

```python
# 네트워크 그래프 그리기
!pip install networkx
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt

# 폰트를 Malgun Gothic으로 설정
plt.rc('font', family='Malgun Gothic')

# 빈 그래프 객체 생성
G = nx.Graph()

# df['items']의 엣지를 그래프에 추가
ar = df['items']
G.add_edges_from(ar)

# PageRank 계산
pr = nx.pagerank(G)

# 노드 크기 설정
nsize = np.array([temp for temp in pr.values()])
nsize = 2000 * (nsize - min(nsize)) / (max(nsize) - min(nsize))

# Kamada-Kawai 레이아웃 설정
pos = nx.kamada_kawai_layout(G)

# 그래프 그리기
nx.draw(G, pos, node_size=nsize, with_labels=True, font_family='Malgun Gothic')
plt.show()
```
