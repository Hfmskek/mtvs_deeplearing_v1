```python
# 라이브러리 불러오기
import pandas as pd
```

```python
# 데이터 확인해보기
df = pd.read_csv('appreply.csv') # index_col=0
df.head()
```

```python
df = df.iloc[3:]
df
```

```python
df = df.reset_index()
df
```

```python
df = df.drop(columns=['index','Unnamed: 0'])
df
```

```python
# 라이브러리 불러오기
from konlpy.tag import Mecab
import re
from tqdm import tqdm
```

```python
mecab = Mecab(dicpath="c:/mecab/mecab-ko-dic")
```

```python
def tokenize_text(text):
    text = re.sub(r'[^ㄱ-ㅣ가-힣\s]', '', text)

    mecab_morphs = mecab.pos(text)

    words = []
    for word, pos in mecab_morphs:
        if pos in ['VA', 'VV', 'NNG', 'NNP']:
            words.append(word)

    word_str = '을 '.join(words)
    return word_str
```

```python
token_list = []

for temp in tqdm(df['text']):
    token_list.append(tokenize_text(temp))
```

```python
token_list
```

```python
corpus_list = []

for index in range(len(token_list)):
    corpus = token_list[index]
    if len(set(corpus.split())) < 3:
        corpus_list.append(corpus)

corpus_list

for corpus in corpus_list:
    token_list.remove(corpus)


print(len(token_list))
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation

# CountVectorizer를 사용하여 텍스트 데이터를 피처 벡터로 변환
# max_df: 단어가 너무 자주 등장하면 무시 (0.1 이상인 경우 무시)
# max_features: 최대 피처 수는 1000
# min_df: 최소 문서 빈도 (2 이상인 경우 포함)
# ngram_range: 1-그램과 2-그램을 고려
count_vectorizer = CountVectorizer(max_df=0.1, max_features=1000, min_df=2, ngram_range=(1, 2))

# 텍스트 데이터를 피처 벡터로 변환
feat_vect = count_vectorizer.fit_transform(token_list)
print(feat_vect.shape)  # 피처 벡터의 크기 출력
print(count_vectorizer.vocabulary_)  # 피처 이름과 인덱스를 포함한 어휘 출력
#feat_vect.toarray()
```

```python
# Latent Dirichlet Allocation (LDA) 모델 생성
# n_components: 추출할 토픽 수는 5
lda = LatentDirichletAllocation(n_components=5)
lda.fit(feat_vect)  # 피처 벡터에 LDA 모델을 피팅
```

```python
# 첫 번째 토픽의 600번째 피처에 대한 가중치 출력
lda.components_[0][600]
```

```python
# 첫 번째 토픽에서 가장 높은 가중치를 가지는 상위 10개 단어의 인덱스 출력
lda.components_[0].argsort()[::-1][0:10]
```

```python
# 모든 토픽에 대한 가중치 행렬 출력
lda.components_
```

```python
def display_topic(model, feature_names, num_top_words):
    # 모델의 각 토픽에 대해 반복
    for topic_index, topic in enumerate(model.components_):
        print('Topic ', topic_index + 1)  # 현재 토픽 번호 출력
        # 각 토픽에서 단어 가중치를 내림차순으로 정렬하여 인덱스를 반환
        topic_word_indexes = topic.argsort()[::-1]
        # 상위 num_top_words 개의 단어 인덱스를 선택
        top_index = topic_word_indexes[0:num_top_words]
        print(top_index)  # 상위 단어 인덱스 출력
        # 상위 단어 인덱스를 사용하여 실제 단어를 추출
        feature_concat = [feature_names[i] for i in top_index]
        print(feature_concat)  # 상위 단어 출력

# CountVectorizer에서 추출한 피처 이름을 가져옴
feature_names = count_vectorizer.get_feature_names_out()
# LDA 모델을 사용하여 각 토픽에서 상위 10개 단어 출력
display_topic(lda, feature_names, 10)
```

```python
# LDA 모델을 사용하여 각 문서의 토픽 분포를 추정
# feat_vect: 문서-단어 행렬 (CountVectorizer로 변환된 피처 벡터)
# sent_topic: 문서-토픽 행렬, 각 행은 문서별 토픽 분포를 나타냄
sent_topic = lda.transform(feat_vect)
sent_topic  # 문서-토픽 분포 행렬 출력
```

```python
# 문서별 가장 관련성이 높은 토픽과 그 확률을 저장할 리스트 초기화
sent_topic_per_list = []

# 각 문서에 대해 반복
for temp in range(sent_topic.shape[0]):
    # 해당 문서에서 가장 높은 확률을 가진 토픽의 인덱스를 찾음
    sent_topic_list = sent_topic[temp].argmax()
    # 해당 문서에서 가장 높은 확률 값
    topic_per = sent_topic[temp].max()
    # 문서 번호, 가장 높은 확률을 가진 토픽 인덱스, 확률을 리스트에 추가
    sent_topic_per_list.append([temp, sent_topic_list, topic_per])

# 리스트를 데이터프레임으로 변환, 컬럼명은 'no', '토픽번호', '확률'
topic_df = pd.DataFrame(sent_topic_per_list, columns=['no', '토픽번호', '확률'])

# 기존 데이터프레임 df와 문서별 토픽 정보를 조인 (병합)
doc_topic_df = topic_df.join(df)

# 최종 데이터프레임 출력
doc_topic_df
```
