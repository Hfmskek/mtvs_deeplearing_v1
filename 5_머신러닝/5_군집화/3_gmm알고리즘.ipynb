```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import DBSCAN
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 붓꽃 데이터 로드
iris = load_iris()
X = iris.data

# 데이터 스케일링 (DBSCAN은 스케일에 민감)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

# Gaussian Mixture Model 적용
gmm = GaussianMixture(n_components=3, random_state=42)
gmm.fit(X)

# 클러스터 할당
gmm_labels = gmm.predict(X)

# 결과 시각화
plt.scatter(X[:, 0], X[:, 1], c=gmm_labels, cmap='viridis')
plt.title('GMM Clustering on Iris Data')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()

# 클러스터 라벨과 원본 데이터를 포함한 DataFrame
df_gmm = pd.DataFrame(X, columns=iris.feature_names)
df_gmm['cluster'] = gmm_labels

# 데이터프레임 표시
print(df_gmm)
```
