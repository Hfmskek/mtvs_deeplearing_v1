```python
ratings = ratings.drop("timestamp", axis=1)
```

```python
ratings_matrix = ratings.pivot_table('rating', index='userId', columns="movieId")
ratings_matrix.head()
```

```python
# movieId로 movies 와 ratings 데이터 프레임 합치기
rating_movies = pd.merge(ratings, movies, on="movieId")
rating_movies.head()
```

```python
# pivot_table 만들어보기
ratings_matrix = rating_movies.pivot_table('rating', index="userId", columns="title")
ratings_matrix.head()
```

```python
# NaN 데이터를 0으로 바꾸기
ratings_matrix = ratings_matrix.fillna(0)
ratings_matrix.head()
```

```python
## 영화들간의 유사도 출력해보기
ratings_matrix_T = ratings_matrix.transpose()
ratings_matrix_T
```

```python
# 유사도 계산
from sklearn.metrics.pairwise import cosine_similarity

item_sim = cosine_similarity(ratings_matrix_T, ratings_matrix_T)
```

```python
item_sim_df = pd.DataFrame(data=item_sim, index=ratings_matrix.columns, columns=ratings_matrix.columns)
item_sim_df.head()
```

### 추천하는 방법

```python
# 개인들에게 추천하는 영화 평점 예측하기
def predict_rating(ratings_arr, item_sim_arr):

    # 유사도 점수와 평점 점수의 곱의 합 - dot
    ratings_pred = ratings_arr.dot(item_sim_arr)

    # 유사도 점수 총합
    normaliztaion_factor = np.array([np.abs(item_sim_arr).sum(axis=1)])

    # 추천 점수 - 예상 평점
    recommend_rating = ratings_pred / normaliztaion_factor

    return recommend_rating
```

```python
ratings_pred = predict_rating(ratings_matrix.values, item_sim_df.values)
```

```python
# 최종 결과 확인해보기
ratings_pred_matrix = pd.DataFrame(data=ratings_pred, index=ratings_matrix.index, columns=ratings_matrix.columns)
ratings_pred_matrix.head()
```

```python
from sklearn.metrics import mean_squared_error

# 사용자가 평점을 부여한 영화에 대해서 예측 성능평가 구해보기 
def get_mse(pred, actual):

    pred = pred[actual.nonzero()].flatten()

    actual = actual[actual.nonzero()].flatten()

    mse = mean_squared_error(pred, actual)

    return mse
```
