```python
from sklearn.feature_extraction.text import CountVectorizer

count_vect = CountVectorizer(min_df=1, ngram_range=(1,2))
genres_mat = count_vect.fit_transform(movie_df['genres'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

genres_sim = cosine_similarity(genres_mat, genres_mat)
genres_sim[:2]
```

```python
genres_sim.argsort()[:, ::-1]
```

```python
genres_sim_sorted_ind = genres_sim.argsort()[:, ::-1]
print(genres_sim_sorted_ind[:1])
```

```python
def find_sim_move(df, sorted_ind, title_name, top_n=10):

    title_movie = df[df["title"] == title_name]

    title_index = title_movie.index.values
    similar_indexes = sorted_ind[title_index, : top_n]

    print(similar_indexes)
    similar_indexes = similar_indexes.reshape(-1)

    return df.iloc[similar_indexes]
```

```python
simliar_movies = find_sim_move(movie_df, genres_sim_sorted_ind, "The Godfather", 10)
simliar_movies[['title', 'vote_average']]
```

```python
simliar_movies[['title', 'vote_average']].sort_values(by="vote_average", ascending=False)[:3]
```

```python
percetile = 0.6
m = movie_df['vote_count'].quantile(percetile)
c = movie_df['vote_average'].quantile(percetile)
m, c
```

```python
def weight_vote_average(record):
    v = record['vote_count']
    R = record['vote_average']

    score = ( (v/(v+m))*R ) + ((m/(m+v))*c)

    return score
```

```python
movie_df['weight_vote_average'] = movie_df.apply(weight_vote_average, axis=1)
```

```python
movie_df[['title', 'vote_average','weight_vote_average', 'vote_count']].sort_values(by="weight_vote_average", ascending=False)[:10]
```

```python
def find_sim_movie(df, sorted_ind, title_name, top_n=10):
    title_movie = df[df['title'] == title_name]
    title_index = title_movie.index.values
    
    # top_n의 2배에 해당하는 쟝르 유사성이 높은 index 추출 
    similar_indexes = sorted_ind[title_index, :(top_n*2)]
    similar_indexes = similar_indexes.reshape(-1)
# 기준 영화 index는 제외
    similar_indexes = similar_indexes[similar_indexes != title_index]
    
    # top_n의 2배에 해당하는 후보군에서 weighted_vote 높은 순으로 top_n 만큼 추출 
    return df.iloc[similar_indexes].sort_values('weight_vote_average', ascending=False)[:top_n]

similar_movies = find_sim_movie(movie_df, genres_sim_sorted_ind, 'The Godfather',10)
similar_movies[['title', 'vote_average', 'weight_vote_average']]
```
