```python
plt.title('사용자 현황', size=20)
plt.plot([1,2,3,4], [2,4,6,2], # x축 데이터와 y축 데이터로 선 그래프 그리기
         c='r', # 선의 색상을 빨간색으로 설정
         lw=4, # 선의 두께를 4 픽셀로 설정
         ls='--', # 선의 스타일을 점선으로 설정
         marker='o', # 데이터 포인트에 원형 마커 사용
         ms=8, # 마커의 크기를 8 픽셀로 설정
         mec='g', # 마커의 테두리 색상을 초록색으로 설정
         mew=3) # 마커의 테두리 두께를 3 픽셀로 설정

plt.xticks([1,2,3,4], ['1주','2주','3주','4주'])
plt.xlabel('주')
plt.ylabel('단위 : (1000명)', rotation=90)
```

grid와 범례

```python
plt.title("매출 비교 그래프", size=20)
plt.plot([2,7,5,2], c='r', label="빽다방")
plt.plot([2,6,14,17], c='b', label="스타벅스")
plt.grid(True)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,6))
plt.title("매출 비교 그래프", size=20)
plt.plot([2,7,5,2], c='r', label="빽다방")
plt.plot([2,6,14,17], c='b', label="스타벅스")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(5,6))
plt.subplot(2,1,1)
plt.title("매출 비교 그래프", size=20)
plt.plot([2,7,5,2], c='r', label="빽다방")
plt.legend()

plt.subplot(2,1,2)
plt.plot([2,6,14,17], c='b', label="스타벅스")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(6,3))
plt.subplot(1,2,1)
plt.title("매출 비교 그래프", size=20)
plt.plot([2,7,5,2], c='r', label="빽다방")
plt.legend()

plt.subplot(1,2,2)
plt.plot([2,6,14,17], c='b', label="스타벅스")
plt.legend()
plt.show()
```

```python
plt.title('매장별 매출 데이터')
x = [0,1,2]
y = [10,5,30]
x_names = ['강남구','관악구','서초구']
plt.bar(x, y)
plt.xticks(x, x_names)
plt.xlabel('지역명')
plt.ylabel('매출(만원)')
plt.show()
```

```python
plt.title('매장별 매출 데이터')
y_name = ['서울', '부산', '광주', '인천']
y_pos = [0,1,2,3]
data = [20, 30, 10, 60]
plt.barh(y_pos, data, color=['y','r','b','g'])
plt.yticks(y_pos, y_name)
plt.show()
```

# 파이차트

```python
labels = ['서울','부산','광주','인천']
data = [30,30,50,20]
colors=['y','r','b','g']
explode = [0.1,0,0,0]
plt.pie(data, 
        explode=explode,
        labels=labels,
        colors=colors,
        autopct="%1.1f%%",
        startangle=30
        )
plt.show()
```

```python
import seaborn as sns
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
iris.head()
```

```python
x = iris.petal_length.values
sns.rugplot(x)
plt.show()
```

```python
sns.countplot(x='class', data=titanic)
plt.title('타이타닉 클래스별 승객수')
plt.show()
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris)
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.pairplot(iris)
plt.show()
```

```python
sns.barplot(x='day', y='tip', data=tips)
plt.show()
```

```python
flights.head()
```

```python
df = flights.pivot(index='month', columns='year', values='passengers')
sns.heatmap(df, annot=True, fmt='d')
plt.show()
```

```python
sns.pointplot(x='day', y='tip', data=tips)
plt.show()
```

```python
plt.title('요일별 전체 매출')
sns.boxplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
plt.title('요일별 전체 매출')
sns.violinplot(x='day', y='total_bill', data=tips)
plt.show()
```

```python
plt.title('요일별 전체 매출')
sns.stripplot(x='day', y='total_bill', 
              data=tips, jitter=True)
plt.show()
```

```python
plt.title('요일별 전체 매출')
sns.swarmplot(x='day', y='total_bill', data=tips)
plt.show()
```
