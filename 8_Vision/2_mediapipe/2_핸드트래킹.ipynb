### 1. 기본 핸드 트래킹

```python
import cv2
import mediapipe as mp
```

```python
# 얼굴 감지 객체와 그리기 객체 초기화
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
drawing_spec1 = mp_drawing.DrawingSpec(thickness=4, circle_radius=4, color=(0, 255, 0))
drawing_spec2 = mp_drawing.DrawingSpec(thickness=2, color=(0, 0, 255))

# 웹캠 설정
cap = cv2.VideoCapture(0)

with mp_hands.Hands(
    max_num_hands = 2, # 트래킹할 최대 손의 수
    min_detection_confidence=0.5, # 손 검출 최소 신뢰
    min_tracking_confidence=0.5 # 손 트래킹을 위한 최소 신뢰
    ) as hands:

    while cap.isOpened():
        res, image = cap.read()
        if not res:
            print('웹캠에서 이미지를 가져오는 것을 실패')
            break
        
        # bgr 이미지를 rgb 변환
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # 속도 개선을 위한 이미지 쓰기 불가능
        image.flags.writeable = False

        # 손감지
        results = hands.process(image)

        # 이미지 다시 쓰기로 지정
        image.flags.writeable = True

        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        if results.multi_hand_landmarks:
            for hand_landmark in results.multi_hand_landmarks:
                mp_drawing.draw_landmarks(
                    image, hand_landmark, mp_hands.HAND_CONNECTIONS,
                    drawing_spec1,
                    drawing_spec2
                )
        image = cv2.flip(image, 1)
        cv2.imshow("frame", image)
        if cv2.waitKey(1) == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()

```

```python
# 얼굴 감지 객체와 그리기 객체 초기화
mp_hands = mp.solutions.hands
mp_drawing = mp.solutions.drawing_utils
drawing_spec1 = mp_drawing.DrawingSpec(thickness=4, circle_radius=4, color=(0, 255, 0))
drawing_spec2 = mp_drawing.DrawingSpec(thickness=2, color=(0, 0, 255))

# 웹캠 설정
cap = cv2.VideoCapture(0)

# 가위바위보 판단
def get_hand_pose(hand_landmarks):
    # 손가락 끝과 PIP(손가락 중간 관절) 사이의 거리를 비교하여 손가락이 펴져 있는지 확인
    open_fingers = []
    for i in [4, 8, 12, 16, 20]:  # 엄지부터 새끼손가락까지의 끝 랜드마크 인덱스
        tip = hand_landmarks.landmark[i]  # 손가락 끝
        pip = hand_landmarks.landmark[i - 2]  # 손가락 중간 관절

        # if handedness.classification[0].label == "Right":  # 오른손인 경우 판별하기

        # 엄지는 x좌표를, 나머지 손가락은 y좌표를 사용하여 개폐 상태 확인 - 오른손 기준
        if i == 4:  # 엄지손가락인 경우
            open_fingers.append(tip.x > pip.x)
        else:
            open_fingers.append(tip.y < pip.y)
    
    # '가위', '바위', '보' 판별
    if open_fingers.count(True) == 5:
        return "보"
    elif open_fingers.count(True) == 2 and open_fingers[1]:
        return "가위"
    elif open_fingers.count(True) == 0:
        return "바위"
    else:
        return "모름"
    
    
with mp_hands.Hands(
    max_num_hands = 1, # 트래킹할 최대 손의 수
    min_detection_confidence=0.5, # 손 검출 최소 신뢰
    min_tracking_confidence=0.5 # 손 트래킹을 위한 최소 신뢰
    ) as hands:

    while cap.isOpened():
        res, image = cap.read()
        if not res:
            print('웹캠에서 이미지를 가져오는 것을 실패')
            break
        
        # bgr 이미지를 rgb 변환
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        # 속도 개선을 위한 이미지 쓰기 불가능
        image.flags.writeable = False

        # 손감지
        results = hands.process(image)

        # 이미지 다시 쓰기로 지정
        image.flags.writeable = True

        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        if results.multi_hand_landmarks:
            for hand_landmark, handedness in zip(results.multi_hand_landmarks, results.multi_handedness):
                mp_drawing.draw_landmarks(
                    image, hand_landmark, mp_hands.HAND_CONNECTIONS,
                    drawing_spec1,
                    drawing_spec2
                )
                label = handedness.classification[0].label  # "Right" or "Left"
                print(label)
                print(get_hand_pose(hand_landmark))

        image = cv2.flip(image, 1)
        cv2.imshow("frame", image)
        if cv2.waitKey(1) == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
