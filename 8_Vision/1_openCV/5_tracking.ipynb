```python
# 1. 모듈 불러오기
import cv2
import numpy as np
```

```python
# 2. 카메라 체크
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("카메라가 안 되요")
```

```python
# 3. 트래킹 설정
# - csrt, kcf, mil 등등
tracker = cv2.TrackerCSRT_create()
```

```python
ret, frame = cap.read()
if not ret:
    print("카메라가 안 읽힙니다")
    exit()

rc = cv2.selectROI('frame', frame)
tracker.init(frame, rc)
cv2.destroyAllWindows()
```

```python
while True:
    ret, frame = cap.read()

    if not ret:
        print("카메라 확인해주세요")
        break
    ret, rc = tracker.update(frame)
    print(rc) # 현재 포지션

    cv2.rectangle(frame, rc, (0,0,255),2)
    cv2.imshow("frame",frame)

    if cv2.waitKey(1) == ord('q'):
        break

cv2. destroyAllWindows()
```

```python
# kcf 트래킹, mil트래킹

tracker = cv2.TrackerKCF_create()

ret, frame = cap.read()
if not ret:
    print('카메라가 안됩니다 확인해주세요')
    exit()

rc = cv2.selectROI('frame', frame)
tracker.init(frame, rc)
cv2.destroyAllWindows()

while True:
    ret, frame = cap.read()

    if not ret:
        print("카메라 확인해주세요")
        break
    ret, rc = tracker.update(frame)
    print(rc) # 현재 포지션
    
    cv2.rectangle(frame, rc, (0, 0, 255), 2)
    cv2.imshow("frame", frame)

    if cv2.waitKey(1) == ord('q'):
        break
cv2.destroyAllWindows()
```
